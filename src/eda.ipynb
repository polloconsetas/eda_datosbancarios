```python
# Eliminar la columna "Unnamed: 0" que no aporta información útil
df_csv.drop(columns=['unnamed: 0'], inplace=True)
```

```python
# Lista de columnas que tienen comas en valores numéricos
numeric_columns = ['cons_price_idx', 'cons_conf_idx', 'euribor3m']

# Reemplazar ',' por '.' y convertir a float
for col in numeric_columns:
    df_csv[col] = df_csv[col].astype(str).str.replace(',', '.').astype(float)

# Verificar cambios
df_csv[numeric_columns].head()
```

### Convertir A float o Int Las Columnas Necesarias

### Convertir Formato Fecha

```python
# Asegurarse de que la columna "date" es de tipo string antes de reemplazar
df_csv['date'] = df_csv['date'].astype(str)
```

```python
# Diccionario para reemplazar meses en español por su número correspondiente
meses_a_numero = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}

# Aplicar reemplazo de nombres de meses por números
for mes_es, mes_num in meses_a_numero.items():
    df_csv['date'] = df_csv['date'].str.replace(mes_es, mes_num, regex=True)

# Verificar si el reemplazo fue exitoso
df_csv[['date']].head(10)
```

```python
# Separar los componentes de la fecha
df_csv[['day', 'month', 'year']] = df_csv['date'].str.split('-', expand=True)

# Crear una nueva columna con el formato correcto (YYYY-MM-DD)
df_csv['formatted_date'] = df_csv['year'] + '-' + df_csv['month'] + '-' + df_csv['day']

# Convertir finalmente a datetime
df_csv['formatted_date'] = pd.to_datetime(df_csv['formatted_date'], format='%Y-%m-%d', errors='coerce')

# Verificar los cambios
df_csv[['date', 'formatted_date']].head(10)
```

```python
# Extraer el año
df_csv['year'] = df_csv['formatted_date'].dt.year

# Extraer el mes
df_csv['month'] = df_csv['formatted_date'].dt.month

# Extraer el día de la semana (0=Lunes, 6=Domingo)
df_csv['day_of_week'] = df_csv['formatted_date'].dt.dayofweek

# Extraer el nombre del día de la semana
df_csv['day_name'] = df_csv['formatted_date'].dt.day_name()

# Verificar los cambios
df_csv[['formatted_date', 'year', 'month', 'day_of_week', 'day_name']].head(10)
```

```python
# Convertir "year" y "month" a enteros para eliminar ".0"
df_csv['year'] = df_csv['year'].astype('Int64')  # Maneja valores nulos correctamente
df_csv['month'] = df_csv['month'].astype('Int64')
df_csv['day_of_week'] = df_csv['day_of_week'].astype('Int64')

# Verificar los cambios
df_csv[['year', 'month', 'day_of_week']].head(10)
```

```python
# Eliminar las columnas "date" y "formatted_date"
df_csv.drop(columns=['date', 'formatted_date'], inplace=True)

# Verificar los cambios
df_csv.head()
```

```python
# Verificar valores nulos en el dataset
df_csv.isnull().sum()

# Imputar valores faltantes en "default" (asumir que los nulos significan "No")
df_csv['default'].fillna(0, inplace=True)

# Verificar que los valores nulos se han corregido
df_csv.isnull().sum()
```

```python
# Definir los rangos de edad y etiquetas
bins = [18, 30, 45, 60, 100]  # Rango de edades
labels = ['Joven', 'Adulto', 'Mayor', 'Senior']  # Nombres de los grupos

# Crear nueva columna categórica "age_group"
df_csv['age_group'] = pd.cut(df_csv['age'], bins=bins, labels=labels, right=False)

# Verificar los cambios
df_csv[['age', 'age_group']].head()
```

```python
# Categorizar la duración de la llamada en 3 grupos
df_csv['duration_category'] = pd.cut(df_csv['duration'],
                                 bins=[0, 100, 300, float('inf')],
                                 labels=['Corta', 'Media', 'Larga'],
                                 right=False)

# Verificar los cambios
df_csv[['duration', 'duration_category']].head()
```

### Análisis y Mejoras de los Tipos de Datos

```python
# 1️⃣ Convertir columnas a enteros donde sea necesario
df_csv['age'] = df_csv['age'].astype('Int64')
df_csv['default'] = df_csv['default'].astype('Int64')
df_csv['housing'] = df_csv['housing'].astype('Int64')
df_csv['loan'] = df_csv['loan'].astype('Int64')
```

### Configurar columna "nr_employed"

```python
# 3️⃣ Ver algunos valores únicos en la columna "nr_employed"
df_csv['nr_employed'].unique()
```

```python
# Reemplazar comas por puntos en "nr_employed" y convertir a float
df_csv['nr_employed'] = df_csv['nr_employed'].str.replace(',', '.').astype(float)

# Verificar el cambio
df_csv[['nr_employed']].head()
```

```python
# 4️⃣ Imputar valores faltantes
df_csv['age'] = df_csv['age'].fillna(df_csv['age'].median())
df_csv['job'] = df_csv['job'].fillna('Desconocido')
df_csv['marital'] = df_csv['marital'].fillna(df_csv['marital'].mode()[0])
df_csv['education'] = df_csv['education'].fillna('Desconocido')
df_csv['housing'] = df_csv['housing'].fillna(0)
df_csv['loan'] = df_csv['loan'].fillna(0)
df_csv['cons_price_idx'] = df_csv['cons_price_idx'].fillna(df_csv['cons_price_idx'].mean())
df_csv['euribor3m'] = df_csv['euribor3m'].fillna(df_csv['euribor3m'].mean())
df_csv['day_of_week'] = df_csv['day_of_week'].fillna(df_csv['day_of_week'].mode()[0])
```

### Gestionar Nulos
